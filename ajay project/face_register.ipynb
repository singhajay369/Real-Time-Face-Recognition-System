```python
import cv2
import face_recognition as fr
import pandas as pd

# Load face detection classifier
fd = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_alt.xml")
filename = "database.csv"

# Load existing face data
try:
    face_db = pd.read_csv(filename, index_col=0)
    data = {"name": face_db["name"].values.tolist(), "enc": face_db["enc"].values.tolist()}
except Exception as e:
    print(e)
    data = {"name": [], "enc": []}

name = data["name"]
enc = data["enc"]

# Function to register face
def register_face():
    vid = cv2.VideoCapture(0)
    captured = True
    counter = 0

    while captured:
        flag, img = vid.read()

        if flag:  # if flag is true then only show image
            faces = fd.detectMultiScale(img, 1.1, 5)
            for x, y, w, h in faces:
                cv2.rectangle(img, pt1=(x, y), pt2=(x + w, y + h), color=(0, 255, 0), thickness=3)
                cropimg = img[y:y + h, x:x + w].copy()

                if len(faces) == 1:
                    counter += 1
                    print(counter)
                    if counter == 20:
                        user_name = input("Enter your name: ")
                        name.append(user_name)
                        encoding = fr.face_encodings(cropimg)
                        if encoding:
                            enc.append(encoding[0].tolist())
                            data = {"name": name, "enc": enc}
                            face_db = pd.DataFrame(data)
                            face_db.to_csv(filename)
                            captured = False
                else:
                    counter = 0

            cv2.imshow("Webcam image", img)
            key = cv2.waitKey(1)
            if key == ord("q"):
                break
        else:
            break

    cv2.destroyAllWindows()
    vid.release()

if __name__ == "__main__":
    register_face()
```
