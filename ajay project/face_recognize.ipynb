```python
import cv2
import pandas as pd
import face_recognition as fr
import numpy as np
import ast

# Load face detection classifier
fd = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_alt.xml")
filename = "database.csv"

# Load existing face data
try:
    face_db = pd.read_csv(filename, index_col=0)
    data = {"name": face_db["name"].values.tolist(), "enc": face_db["enc"].values.tolist()}
except Exception as e:
    print(e)
    data = {"name": [], "enc": []}

name = data["name"]
enc = data["enc"]

# Function to recognize face
def recognize_face():
    vid = cv2.VideoCapture(0)
    captured = True

    while captured:
        flag, img = vid.read()
        cropped_face = img

        if flag:
            faces = fd.detectMultiScale(img, 1.1, 5)
            for x, y, w, h in faces:
                cropped_face = img[y:y + h, x:x + w].copy()

                if len(faces) == 1:
                    fresh_face_enc = fr.face_encodings(cropped_face)
                    for ind, fe in enumerate(enc):
                        try:
                            # Safely evaluate the string representation of the list
                            fe_list = ast.literal_eval(fe)
                            matched = fr.compare_faces(fresh_face_enc, np.array(fe_list))[0]

                            if matched:
                                print("Matched")
                                print(name[ind])
                                cv2.putText(img, name[ind], (x, y), cv2.FONT_HERSHEY_COMPLEX, 1.5, (0, 0, 255), 4)
                                cv2.rectangle(img, pt1=(x, y), pt2=(x + w, y + h), color=(0, 255, 255), thickness=2)
                            else:
                                print("Not matched")

                        except IndexError:
                            print("No face is detected!!")

            cv2.imshow("Webcam image", img)
            key = cv2.waitKey(1)
            if key == ord("q"):
                break
        else:
            break

    cv2.destroyAllWindows()
    vid.release()

if __name__ == "__main__":
    recognize_face()
```
